```python
# FUNCION PARA LA CARGA  DE DATOS EMIGRANTES
def cargar_datos_desde_url(url):
    try:
        datos = pd.read_html(url)
        num_tablas = len(datos)
        if num_tablas > 0:
          pass
        return datos, num_tablas
    except Exception as e:
        return None, 0
```

```python

#FUNCION DE EXTRACCION DE DATOS Y ASIGNAR EL PAIS POR DATAFRAME
def emigrantes_pais(lista_strin):
    k=[]
    for i in lista_strin:
      url_emigrantes = i #
      datos, num_tablas = cargar_datos_desde_url(url_emigrantes)
      poblacion_emigrante_country = datos[0]
      poblacion_emigrante_country['pais']=i.split('/')[6].replace('-',' ')
      k.append(poblacion_emigrante_country)
    return k
```
